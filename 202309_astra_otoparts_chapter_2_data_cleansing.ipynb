```python
# There are weird values with NAN, let's fix them
data['director_name'] = np.where(data['director_name']=='N/A','', data['director_name'])
data['director_name'] = np.where(data['director_name']=='Nan','', data['director_name'])
data['director_name'] = np.where(data['director_name']=='Null','', data['director_name'])
```

# Filling missing data

```python
data['gross']
```

```python
import seaborn as sns

sns.distplot(data['gross'])
```

```python
sns.distplot(data['gross'].fillna(data['gross'].mean()))
```

```python
sns.distplot(data['gross'].fillna(data['gross'].median()))
```

```python
sns.distplot(data['gross'].fillna(data['gross'].median()), label='median')
sns.distplot(data['gross'].fillna(data['gross'].mean()), label='mean')
plt.legend()
plt.show()
```
